```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel
import torch
```

```python
# Load the base model (Open LLaMA 3B model)
base_model = AutoModelForCausalLM.from_pretrained("openlm-research/open_llama_3b_v2")

# Load the fine-tuned model
model = PeftModel.from_pretrained(base_model, "kushagragoyal/fine_tuned_model")
```

```python
# Load the tokenizer for the base model
tokenizer = AutoTokenizer.from_pretrained("openlm-research/open_llama_3b_v2")
```

```python
# Set the model to evaluation mode
model.eval()
```

```python
# Define your input text (the prompt)
input_text = "make directory and add readme inside it just this nothing more"
```

```python
# Tokenize the input text
input_ids = tokenizer.encode(input_text, return_tensors="pt")
```

```python
# Generate output (text) using the model
with torch.no_grad():  # No need to track gradients during inference
    outputs = model.generate(input_ids, max_length=100, num_return_sequences=1)
```

```python
# Decode the generated output into text
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

# Print the generated text
print(generated_text)
```
